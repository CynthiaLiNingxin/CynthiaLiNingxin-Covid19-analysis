```python
import pandas as pd
import numpy as np
import qeds
import datetime
```

```python
#Import data from csv file.
df = pd.read_csv(r'G:\UTSG\2022 Winter\ECO225\Project\dataset\csse_covid_19_data\csse_covid_19_time_series\time_series_covid19_confirmed_global.csv', index_col=0)
df
```

```python
#Drop column that we don't need.
df.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
#Some of the columns' name have error, filt them and rename them.
for label in df.columns:
    if '2' in label:
        if '200' in label:
            new_label = label[3:]
            df.rename(columns={label: pd.to_datetime(new_label)}, inplace=True)
        else:
            df.rename(columns={label: pd.to_datetime(label)}, inplace=True)
df
```

```python
all_countries = df.groupby('Country/Region').sum()
all_countries
```
